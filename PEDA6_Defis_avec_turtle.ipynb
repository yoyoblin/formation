### 5. Figure à réaliser : a1.png
---

> Modifiez et complétez le programme suivant :

```python
from turtle import * 
speed(4)

for i in range(2) :
    left(90)
    forward(80)
    right(90)
    forward(80)


done() # 
```

### 6. Figure à réaliser à partir du parallélogramme dessiné dans le programme ci-dessous a2.png
---

> A quoi sert le bloc d'instruction **def parallelogramme()** ?
> Modifiez et complétez le programme suivant :

```python
from turtle import * 
speed(4)

def parallelogramme () :
    for i in range(2) :
        color("red")
        forward(80)
        left(40)
        forward(60)
        left(140)

for i in range(3) :
    parallelogramme()
        
done() # 
```
