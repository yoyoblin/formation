## **2. Exemples d'utilisations :**


---




- **La vérité est universelle...**

```python
if True :
  print ("Ok alors..") # par défaut la mémoire booléenne est réglée sur True
```

```python
if False :
  print ("bizarre ?") # et rien ne s'écrit...  la vérité est ailleurs
```

- **Avec uniquement deux conditions :** on emploie if et else

> Bloc en retrait



```python
n = int(input("entrez un nombre entier"))

if n%2 == 0:
    print("ce nombre est pair")
else:
    print("ce nombre est impair")
```

> ...cela fonctionne avec des intervalles :

```python
n = int(input("entrez un nombre entier"))

if 0 <= n <= 5 :
  print("entre 0 et 5")

else:
  print("supérieur à 5")
```

> cela fonctionne avec des str, ici une version "raccourcie" (ne pas oublier le ":")

```python
mot = "pomme"

if "p" in mot : print ("il y en a aussi")
else : print ("C'est curieux chez les marins ce besoin de faire des phrases")
```

- **Avec plusieurs conditions :** emploi de if, elif, else

```python
from math import *

a, b, c = 3, 5, 1 # affectation façon "tuple"

d = b ** 2 - 4 * a * c  # Vous le reconnaissez
print ("delta = ", d)

if d > 0 :
    x1 = round((- b + sqrt (d))/( 2 * a),2)
    x2 = round((- b - sqrt (d))/( 2 * a),2)
    print("deux solutions", x1,"  ",x2)

elif d == 0 :
    x = round( - b /( 2 * a),2)
    print("une seule solution = ", x)

else :
    print("pas de solution réelle")
```

> Exemple parce que c'est le moment :

```python
revenu = int(input("Entrez votre revenu imposable : "))
nombre = int(input("Entrez le nombre de parts : "))
revenu_imp = revenu / nombre
print("Votre revenu imposable est de ",revenu_imp, " €")

if revenu_imp <= 9964 :
    impot = 0

elif revenu_imp <= 27519 :
    impot = revenu * 0.14 - 1394.96 * nombre

elif revenu_imp <= 73779 :
    impot = revenu * 0.30 - 5798 * nombre

elif revenu_imp <= 156244 :
    impot = revenu * 0.41 - 13913.69 * nombre

else :
    impot = revenu * 0.45 - 20163.45 * nombre
        
impot = round(impot,2)
print("Votre impôt s'élève à :", impot, " €" )
```

> Une structure conditionnelle si elle est répétitive trouve sa place dans une fonction python :

```python
liste = [1.56,1.72,1.60,1.84,1.66,1.78,1.64,1.82,1.72,1.65,1.85,1.62]

def med(liste) :
  liste.sort()     # tri dans l'ordre'
  n = len(liste)   # nombre de valeurs de la liste
  if n%2 == 1 :    # si n est impair
    num = int((n+1)/2)
    return liste[num-1] # trouve le nombre corresondant à ce numéro

  else :           # sinon si n est pair
    nombre1 = liste[int(n/2 - 1)]
    nombre2 = liste[int(n/2)]
    return (nombre1 + nombre2)/2  

def q1(liste) :
  liste.sort()
  n = len(liste)
  num = int(n/4) + 1  # quel est le numéro du quartile 1 ?
  return liste[num-1]   # trouve le nombre corresondant à ce numéro

def q3(liste) :
  liste.sort()
  n = len(liste)
  return liste[int(3*n/4)] 

print ("médiane =",med(liste))
print ("quartile 1 =",q1(liste))
print ("quartile 3 =",q3(liste))
```
