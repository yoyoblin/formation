exemple avec $$ f(x)=x^{2}$$

```python
x =float(input("x= ")) # entrez des valeurs de type float
fx = x**2

print(x,"-->",fx)
```

>>Méthode par balayage avec une boucle for : cible c =7

```python
for x in range(0,5) : # liste d'entiers, pas de valeurs décimales 
  fx = x**2
  print (x,"-->",fx)
```

>> Modifiez le programme pour encadrer la solution à 0,01 près

>> Variante avec balayage automatique dans une boucle while

```python
x= 2
fx = 0
while abs(fx-7)> 0.1 : # attention à la précision demandée et à l'incrément
   fx = x**2
   print (x,"-->",fx)
   x = x + 0.01
```

> Même si ici les fonctions python n'apportent rien on peut quand même les présenter

```python
x = 12 # initialisation de la variable

# les fonctions sont définies avant la boucle (du coup inutile d'initialiser des variables)
def op1(x) :
  return round(99.90 + 19.99*x,2)
def op2(x) :
  return round(24.90*x,2)

while op1(x) >= op2(x)  : 
  print (x," ",op1(x)," ",op2(x))
  x = x+1

print ("si je choisis l'operateur 1, il faut rester engagé au moins",x,"mois")
```
