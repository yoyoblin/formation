## **Problème 2 : Résolution d'équations** (Réflexion algorithmique, Modification) $$f(x) = c$$


---




>**Exemple avec la fonction carré.** Utilisez le programme suivant pour trouver à 0,1 près quel nombre a pour image 7



```python
x =float(input("x= ")) # on saisit des valeurs de type float
fx = x**2

print(x,"-->",fx)
```

> **Pour aller plus vite** vous aller calculer les images des nombres entiers de l'intervalle **[Ø ; 5[** avec une boucle **FOR** : l'inconvénient c'est qu'on ne peut pas utiliser des nombres à virgule.

```python
for x in range(0,5) : # liste d'entiers, pas de valeurs décimales 
  fx = x**2
  print (x,"-->",fx)
```

> **Modifiez le programme** pour encadrer la solution à 0,01 près.  

```python
from numpy import * 

for x in arange(2,3,0.1) : # permet de créer des liste arange() avec un pas décimal
  fx = round(x**2,5) 
  print (x,"-->",fx)
```

> **Variante avec balayage automatique dans une boucle WHILE** : attention il y a un risque de créer une boucle infinie, Pas forcément pratique...

```python
x= 2
fx = 0
while abs(fx-7)> 0.1 : # attention à la précision demandée et à l'incrément
   fx = x**2
   print (x,"-->",fx)
   x = x + 0.01
```

> - On peut aussi calculer tous les tarifs **tant que** le premier est plus cher avec une boucle **WHILE** et des **fonctions**.

```python
x = 12 # initialisation de la variable

def op1(x) :
  return round(99.90 + 19.99*x,2)
def op2(x) :
  return round(24.90*x,2)

while op1(x) >= op2(x)  : 
  print (x," ",op1(x)," ",op2(x))
  x = x+1

print ("si je choisis l'operateur 1, il faut rester engagé au moins",x,"mois")
```
