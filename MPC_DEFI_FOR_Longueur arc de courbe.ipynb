## <span style="color:orange"> A. Parenthse sur les tracs de courbes avec `numpy` et `matplotlib.pyplot`

>- Les tableaux **arange()** et **linspace()** de Numpy n'ont pas les limitations de **range()** et apportent un peu de "complexit utile".

```python
from numpy import *
x = arange(0,5,0.5)  # des valeurs de 0  5 (exclu) avec un pas de 0,5

print(x)
```

```python
from numpy import *
x = linspace(0,5,9) # 9 valeurs de 0  5 rparties uniformment

print(x)
```

>- **NumPy** permet d'appliquer directement des calculs ou des fonctions  un tableau : ces calculs sont appliqus  chacun des lments du tableau en crant un tableau de rsultats (voir ci-dessous "y").
___
> - **Exemple de trac de courbes avec numpy et mathplotlib:** courbe "lisse"

```python
import numpy as np      # les modules sont imports par prfixes pour bien distinguer leurs fonctions
import matplotlib.pyplot as plt
plt.clf() # efface la mmoire graphique

x = np.linspace(0, 5) # Dfinit un tableau de valeurs de 0  5 (tableau 1D Numpy)
y = x ** 2            # Cre un tableau d'images des nombres de la liste prcdente

plt.plot(x, y) # Trace la courbe  partir des listes prcdentes
plt.axis([0,5,0,25]) # zone de trac
plt.title("courbe")
plt.ylabel("axe des ordonnes")
plt.xlabel("axe des abcisses")
plt.grid() # quadrillage
plt.show()
```

>- **Instruction `arange()`** pour dfinir un pas qui peut tre dcimal, contrairement  **`range()`**
>- La syntaxe est **`arange(depart,fin(exclu),incrment)`** ce qui permet dobtenir un tableau 1D allant dune valeur de dpart  une valeur de fin (exclue) avec un incrment.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.clf() # efface la mmoire graphique

x = np.arange(0,5.5,0.5)  # Dfinit une liste de valeurs de 0  5 avec un pas de 0.5
y = x**2  

plt.plot(x, y,marker='.')
plt.axis('equal') # repre orthornorm
plt.ylabel("axe des ordonnes")
plt.xlabel("axe des abcisses")
plt.grid()
plt.show()
```

>- **Instruction `linspace()`** pour dfinir le nombre de points.
>- La syntaxe est **`linspace(depart,fin,nbre elements)`** ce qui permet dobtenir un tableau 1D allant dune valeur de dpart  une valeur de fin avec un nombre donn dlments.

```python
import numpy as np      # les modules sont imports par prfixes pour bien distinguer leurs fonctions
import matplotlib.pyplot as plt
plt.clf() # efface la mmoire graphique

x = np.linspace(0,5,4) # Dfinit un tableau de 4 valeurs de 0  5
y = x**2            

plt.plot(x, y,'g',marker='x') # Trace la courbe en vert avec des points
plt.axis('equal') # repre orthornorm
plt.ylabel("axe des ordonnes")
plt.xlabel("axe des abcisses")
plt.grid() # quadrillage
plt.show()
```

```python
import numpy as np      # les modules sont imports par prfixes pour bien distinguer leurs fonctions
import matplotlib.pyplot as plt
plt.clf() # efface la mmoire graphique

x1 = np.linspace(0, 5) 
x2 = np.arange(0,6,1.25)
y1 = x1 ** 2            
y2 = x2 ** 2   

plt.plot(x1, y1, color = 'b') 
plt.plot(x2, y2, color = 'r',marker='o') # trac avec des points
plt.ylabel("axe des ordonnes")
plt.xlabel("axe des abcisses")
plt.grid() # quadrillage
plt.show()
```

## <span style="color:orange"> B. Calcul de l'approximation de la longueur d'arc par trois manires diffrentes

>- **Calcul d'une approximation avec une boucle while :** 
___
>>- <span style="color:blue"> **Si le pas est fix  1 :** 6 points (5 intervalles)

```python
from math import*

# distance entre deux points d'abscisses x1 et x2
def long(x1,x2) :
    y1 = f(x1)
    y2 = f(x2)
    return sqrt((x2-x1)**2 + (y2-y1)**2)

# l'image par la fonction f
def f(x):
    return x**2

l = 0
x = 0

# balayage des valeurs de 0  4 avec un pas de 1 (voir  la fin de la boucle)
while x <= 4 :
    x1 = x
    x2 = x + 1
    
    l = l + long(x1,x2)
    x = x + 1

print (l)
```

>>- <span style="color:blue"> **Si le pas est fix  0.1 :** 51 points (50 intervalles)

```python
from math import*

# distance entre deux points d'abscisses x1 et x2
def long(x1,x2) :
    y1 = f(x1)
    y2 = f(x2)
    return sqrt((x2-x1)**2 + (y2-y1)**2)

# l'image par la fonction f
def f(x):
    return x**2

l = 0
x = 0

# balayage des valeurs de 0  4,9 avec un pas de 0.1 (voir  la fin de la boucle)
while x <= 5 - 0.1 :
    x1 = x
    x2 = x + 0.1
    l = l + long(x1,x2)
    x = x + 0.1

print (l)
```
