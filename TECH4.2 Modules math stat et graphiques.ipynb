## <span style="color:orange"> A. Comptage et fréquence dans une chaîne de caractères ou une liste

> - **Exercice :** modifie le programme pour calculer la fréquence d'une autre lettre

```python
texte = "lalumièreestbleueetlesoiseauxsontblancs" # le texte sans trous

# Comptage du nombre d'apparition de la lettre a
l = len(texte) # nombre de lettres
n = texte.count("a") # comptage de la lettre choisie


# Calcul de la fréquence arrondie à deux chiffres après la virgule
f = round(n/l * 100,2) 
print(l, "lettres dont", n,"a")
print(f,"% de lettre a")
```

> ### 2. Détermination de la médiane par comptage dans une liste

> - **Il faut ordonner la liste et accéder aux valeurs par leur index :**  on utilise `liste.sort()` et `liste[index]`
> - On découvre également les tests python : `if` et `else`
> - le problème majeur est le décalage entre **la numerotation** de la liste et **l'index python**

```python
liste = [2,2,5,8,4,5,6,10,9,10,7,9,1,8,7,6,7,9,9,2,1,10,5,3,5,6,7,4,8,3,4,5,6,5,8,6]

liste.sort() # Trier les nombre dans l'ordre croissant
n = len(liste)

# Si le nombre est impair
if nombre%2 == 1 :  
    numero = int((n + 1)/2)
    mediane = liste[numero-1] # décalage entre numero et index

# Sinon si le nombre est pair
else :   
    numero = int(n/2)
    valeur1 = liste[numero-1] 
    valeur2 = liste[numero]
    mediane = (valeur1 + valeur2)/2
    
print ("la médiane est égale à ",mediane)
```

## <span style="color:orange"> C. Approche "outil" avec l'utilisation des bibliothèques `numpy` , `matplotlib.pyplot` ou `matplotlib.pylab`

> ### 1. Utilisation de numpy
> - Le module **numpy** comporte beaucoup de fonctions utiles, on l'importe complètement avec import * ou on peut l'importer par "préfixe" 
> - Ce module (souvent couplé avec **matplotlib.pyplot** ) permet de créer des graphiques... mais est ce plus interressant que les grapheurs ?

```python
from numpy import *

# voici les notes d'élèves à un devoir
list = [4,3,8,4.5,10,4,6.5,7,1.5,7.5,9,9.5,5,5.5,3.5,7]

minimum = min(list)
maximum = max(list)
ampli = ptp(list)  # peak to peak (amplitude)
moy = round(mean(list),2)# moyenne
q1 = percentile(list,25)
q3 = percentile(list,75)
med = median(list)
ecart = std(list)  # standard deviation (écart type)


print ("le quartile numéro 1 est égal à ",q1)
print ("le quartile numéro 3 est égal à ",q3)
print ("ma médiane est égale à ",med)
print ("ma moyenne est égale à ",moy)
print ("le minimum est égal à ",minimum)
print ("le maximum est égal à ",maximum)
print ("l'amplitude est égale à ",ampli)
print ("l'écart type1 est égal à ",ecart)
```

> - **Exemple** de comparaison de deux séries de valeurs (les boîtes à moustaches viennent après)

```python
from numpy import *

list1 = [10,2,3,1,5,3,5,6,10,2,5,7,9,5]
list2 = [6,2,5,1,4.5,3,5,6,9,7.5,5,3.5,5,4.5]

liste = [list1,list2]

for l in liste :
    print ("moyenne = ", mean(l))
    print ("médiane =", median(l))
    
    print ("amplitude =",max(l)-min(l))
    print ("écart interquartile = ",percentile(l,75)-percentile(l,25))
```

## <span style="color:orange"> D. Tracés de courbes avec l'utilisation des bibliothèques `numpy` , `matplotlib.pyplot` et `scipy`

- **L’utilisation de Numpy permet de réduire le nombre de boucles :** lorsqu’une fonction mathématique est appliquée à une liste Numpy, elle va automatiquement agir sur chaque élément de la liste. Les fonctionnalités peuvent être limitées notamment en ce qui concerne la concaténation **avec des floats**
- **Les tableaux numpy** *numpy.array([ ])* sont plus complets. Ce sont des ensembles de valeurs. Ils permettent des calculs sur directement sur les tableaux avec des nombres décimaux.
- Ces tableaux sont à **plusieurs dimensions :** 1D <--> une seule liste, 2D <--> deux listes, etc... ce sont des matrices.
- **Avec ces deux listes** le module graphique **pyplot** va permettre de tracer la représentation graphique de la fonction.
- Ensuite on peut aussi utiliser **scipy.stats** pour effectuer une régression linéaire : voir exemple en électricité

> - **Tracé de courbes :** (petite parenthèse sur les fonctions)

```python
import numpy as np      # les modules sont importés par préfixes pour bien distinguer leurs fonctions
import matplotlib.pyplot as plt
plt.clf() # efface la mémoire graphique

x = np.linspace(0, 1) # Définit un tableau de valeurs de 0 à 1 (tableau 1D Numpy)
y = x ** 2            # Crée un tableau d'images des nombres de la liste précédente

plt.plot(x, y) # Trace la courbe à partir des listes précédentes
plt.plot(y, x) # inversion des axes
plt.grid() # quadrillage
plt.show()
```

>- Utilisation du module **stats** de **scipy** pour faire une régression linéaire

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sc
plt.clf() # efface la mémoire graphique

# valeurs expérimentales

I = [0.06, 0.09, 0.15, 0.21, 0.26, 0.31]
U = [2.1, 3.4, 4.5, 5.6, 7.5, 8.4]

plt.plot(I, U) # Trace la courbe de U en fonction de I
plt.xlabel("intensité en A")
plt.ylabel("tension en V")
plt.grid() # quadrillage

droite = sc.linregress(I,U) # regression linéaire
coef = droite.slope # coefficient de la droite de régression
print ("la résistance est de ",coef,"ohms")

plt.show()
```

>- <span style="color:red"> Petite subtilité suplémentaire en utilisant les **tableaux numpy**, très pratiques pour tracer des fonctions et des courbes expérimentales `numpy.array([ ])`
>>- **Utilisation de données expérimentales :**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sc
plt.clf() # efface la mémoire graphique


I = np.array([0.06, 0.09, 0.15, 0.21, 0.26, 0.31]) # Crée un tableau numpy, plus complet qu'une liste
U = np.array([2.1, 3.4, 4.3, 5.6, 7.5, 8.1])

plt.plot(I, U) # Trace la courbe de U en fonction de I
plt.xlabel("intensité en A")
plt.ylabel("tension en V")
plt.grid() # quadrillage

droite = sc.linregress(I,U) # regression linéaire
coef = droite.slope # coefficient de la droite de régression
b = droite.intercept # coeff b

U = coef * I + b # gros avavantage des tableaux Numpy : on peut faire des opérations directement
plt.plot(I,U)

plt.show()
```

>>- **Problème de statistiques à deux variables :**

```python
import matplotlib.pylab as plt
import scipy.stats as sc
plt.clf() # efface la mémoire graphique


# tableaux de valeurs
x = plt.array([1400,1600,1850,2250,1800,2000,2590,3150,3000,4000]) 
y = plt.array([4100,5400,2000,4700,3600,4500,5300,3500,5200,6000])

plt.title("chiffre d'affaire/connexions")
plt.xlabel("nombre de connexions")
plt.ylabel("chiffre d'affaire")




# NE RIEN MODIFIER SOUS CETTE LIGNE !!!
# repère
plt.scatter(x, y, c = "green") # Nuage de points bleus
plt.grid() # quadrillage
plt.axis([0,5000,0,8000]) # repère mini/maxi

# ajustement affine
droite = sc.linregress(x,y)
a = droite.slope     # coeff directeur
b = droite.intercept # coeff b
r2 = droite.rvalue**2 # r carré

# tracé de la droite
y2 = a*x+b #tableau Numpy 
plt.plot(x,y2,c = "orange") # tracé de la droite
plt.show()

print ("les coefficients a et b de la droite d'ajustement affine sont les suivants :")
print ("a=",a,"b=",b)
print ("le coefficient de détermination est :", "r2=",r2)
```
