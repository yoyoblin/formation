## <span style="color:orange">2. C'est long ? Vous allez à présent utiliser une boucle pour créer des répétitions. <span>
---

### a) Une boucle crée une répétition : si vous connaissez le nombre de ces répétitions alors utilisez une boucle *for* (on dit que c'est une boucle bornée)



>- ***for n in range (1,6) :*** permet de répéter **5 fois** les mêmes instructions car **n va prendre les valeurs 1, 2, 3, 4, 5**
>- **Les instructions** qui se répètent sont décalées : on dit qu'elles sont **indentées**. 
>- **Exemple ci-dessous :**


```python
for n in range(1,6):  # 5 répétitions (6-1 = 5) car n prend les valeurs 1, 2, 3, 4, 5... et pas 6 !!
    print(n)          # affiche la variable 
```

### a) Comment afficher le numéro (l'indice) du terme de la suite : U1, U2, etc... ?

>- ***print("texte") :*** permet d'afficher du texte.
>- ***print("texte",variable) :*** permet d'afficher du texte et une variable. Les différents éléments sont séparés par des **virgules**;

**Exemple en reprenant le programme précédent :**

suite01.jpg

```python
U = 4

for n in range (1,5) : 
    print ("U",n,"=",U) # les lettres entre guillemets sont du texte
    U = U + 3
```
