## **2. Turtle : les boucles**
---

- **La boucle for** avec une variable de boucle. 
C'est une boucle bornée : Elle répète les instructions un nombre prédéfini de fois.

- **La boucle while** avec une variable incrémentée. 
C'est une boucle non bornée : Elle répète les instructions tant que la condition est vraie.

> ###  La boucle "for"
---
> **Dans les exemples ci-dessous la boucle est bornée :** la variable de boucle prend toutes les valeurs (int ou str) d'un **itérable**, c'est à dire d'une châine de caractère ou d'une liste. 

> Cette variable de boucle **n'est pas forcément utilisée** dans les instructions de la boucle, car ce n'est pas une obligation. De plus et elle n'a pas besoin d'être incrémentée.

> **Exemple d'activité :** Modifiez le programme pour dessiner un autre polygone régulier (pentagone, hexagone..)

```python
from turtle import *
speed(6) 

for i in range(6) :
  forward(100)
  left(60)

done()
```

> **Exemple d'activité :** Modifiez le programme pour dessiner un autre pentagone régulier en changeant les couleurs

```python
from turtle import *
speed(6) 

for i in ["red","green","blue","orange","yellow","grey"] :
  color(i)
  forward(100)
  left(60)

done()
```

> **Remarque :** la variable ici ne sert à rien dans la boucle. Changez le mot en ligne 4 pour comprendre à quoi il sert..

```python
from turtle import *
speed(6) 

for i in "python" :
  forward(100)
  left(60)

done()
```

> ###  La boucle "while"
---
> **Dans les exemples ci-dessous la boucle est non bornée :** elle se répète "tant que la condition est vraie", de plus elle a besoin d'être incrémentée 

```python
from turtle import *
speed(6) 

i = 0 # initialisation obligatoire de la variable compteur
while i < 6 :
  forward(100)
  left(60)
  i = i + 1

done()
```

## **3. Turtle : les fonctions**
---


- **Une fonction** renvoie une variable par l'intermédiaire de **return**
- **Une fonction sans renvoi** est une procédure, un sous programme.



> **Exemple d'activité :** la machine à polygones, la fonction ci-dessous utilise un paramètre associé au nombre de côtés du polygone. 

> Quelle valeur devez vous écrire en ligne 7 à la place du point d'interrogation ?

```python
from turtle import *
speed(6)

def poly (n) :  # cette fonction dessine un polygone régulier à n côtés
    for i in range (n) :
        forward (100)
        left (?/n)

n = int(input("nombre de côtés "))
poly(n)
    
done()
```

> **Juste à titre de démo :** avec deux fonctions différentes, avec et sans paramètres

```python
from turtle import *
from random import * # module aléatoire
speed(4)

def distance(i) :  # cette fonction a un parametre et renvoie une valeur
  return i*10

def angle() :  # cette fonction n'a pas de parametre et renvoie une valeur
  return randint(50,100)

for i in range (0,20) :
  forward(distance(i))
  left(angle())

done()
```
