### A.1.c. Chronométrer le temps d'exécution avec `timeit`

Il est possible de chronométrer le temps d'exécution total de `10 000` appels d'une fonction, avec le module `timeit`. Le résultat est donné en secondes :

````python
from timeit import timeit

def listeZeros(n):
    lst=[]
    for i in range(n):
        lst.append(0)
    return lst

timeit("listeZeros(500)", globals=globals(), number= 10_000)
````

### A.1.d. Visualiser un graphique avec `matplotlib.pyplot`

Il est possible de visualiser un graphique avec `matplotlib.pyplot`.

Par exemple, après avoir exécuté la cellule précédente, on peut comprendre ci-dessous la complexité en O(n) de l'algorithme :

````python
import matplotlib.pyplot as plt

x = list(range(0,50,1))

y = []
for i in x:
    t = timeit("listeZeros("+str(i)+")", globals=globals(), number= 10_000)
    y.append(t)

plt.clf()
plt.title("Temps d'exécution en fonction du paramètre")
plt.plot(x, y)
plt.xlabel('taille de la liste')
plt.ylabel("temps d'exécution (en s)")
plt.show()
````

### A.1.f. Modifier une image avec `PIL`

Il est possible de modifier une image avec `PIL` : 

````python
from PIL import Image
monImage1=Image.open("buse.jpg")#Cette image a une dimension de 400×200 pixels
monImage2=Image.new('RGB',(400,200))#Crée une seconde a une dimension de 400×200 pixels

for x in range(400):
    for y in range(200):
        r,v,b = monImage1.getpixel((x, y)) #r,v,b prennent les valeurs des composante RVB du pixel (x,y) de Image1
        g=(r+v+b)//3                 #calculer ici la valeur de g                  
        monImage2.putpixel( (x, y),(g,g,g))#On fixe la couleur du pxel (x,y) de Image2

monImage4=Image.new("RGB",(806,204))
monImage4.paste(monImage1, (2, 2)  )
monImage4.paste(monImage2, (404, 2))
monImage4                #On affiche une image composée de Image1 et Image2
````

## A.3. Utiliser les métadonnées pour verrouiller des cellules

Il est possible d'empêcher la modification de cellules d'un notebook.

Pour cela, deux méthodes.

### A.3.a. Avec les barres d'outils des cellules

Dans le menu, cliquez sur `Affichage` > `Barre d'outils de cellule` > `Edit Metadata`

Cliquez sur le bouton "Edit Metadata" de la cellule qui vous intéresse puis remplacer le texte JSON (les accolades et leur contenu éventuel) par le suivant, puis cliquez sur le bouton `Edit`.

```json
{
"deletable":false,
"editable":false
}
```

Une fois le travail terminé, `Affichage` > `Barre d'outils de cellule` > `Raw Cell Format`.

### A.3.b. En éditant le fichier `.ipynb`

Vous pouvez à tout moment exporter votre carnet Jupyter en allant dans le menu `Fichier` > `Enregistrer sous...`

Vous obtenez alors un fichier `.ipynb` que vous pouvez ouvrir dans un éditeur de texte.

Le fichier ressemblant au format `JSON` qui s'affiche contient pour chaque cellule un ensemble qui ressemble à cela :


```json
  {
   "cell_type": "code",
   "execution_count": 1,
   "metadata": {},
   "outputs": [],
   "source": [
    "a = 3"
   ]
  },
 ```
 
 
en remplaçant la ligne ` "metadata": {},` par
 
 ```json
  "metadata": {
    "trusted": true,
    "editable": false,
    "deletable": false
  },
  ```
  
la cellule n'est plus modifiable et ne peut plus être supprimée.

## A.4. Tranformer le carnet en diapos interactives avec l'option RISE

Il est possible de tranformer le carnet en **diapos interactive** (les cellules de code restent exécutables) avec l'option **RISE**. 

Pour cela, il faut d'abord spécifier quelles cellules vous souhaitez **regrouper sur une même diapo**, et quelles cellules vous souhaitez rendre visibles dès la diapositive affichée, ou **après l'appui sur la touche <kbd>➡</kbd>**.

Dans le menu, cliquez sur `Affichage` > `Barre d'outils de cellule` > `Slideshow`

Cliquez sur la liste déroulante "Slide type" de la premier cellule, et sélectionnez `Slide`. 

Si vous souhaitez que votre 2ème cellule apparaisse **sur la même diapo** que la première cellule, mais seulement **après l'appui sur la touche <kbd>➡</kbd>**, alors liquez sur la liste déroulante "Slide type" de cette 2ème cellule, et sélectionnez `Fragment`. 

Sinon, si vous souhaitez que votre 2ème cellule apparaisse **sur une nouvelle diapo**, sélectionnez `Slide`.

Sinon, laissez l'option `-` sélectionnée.
    
Répétez cette méthode **pour chaque cellule**.
    
Une fois le travail terminé, `Affichage` > `Barre d'outils de cellule` > `Raw Cell Format`.

Vous pouvez alors cliquer sur l'icône <button class="btn btn-default" title="Enter/Exit RISE Slideshow" id="RISE" data-jupyter-action="RISE:slideshow"><i class="fa-bar-chart fa"></i></button> présente dans la barre d'outils.
    
*Une fois dans le diaporama, vous pouvez cacher les icones en appuyant sur la touche <kbd>,</kbd> et mettre le navigateur en plein écran avec la touche <kbd>F11</kbd>.*
