## <span style="color:orange">2. C'est long ? Vous allez à présent utiliser une boucle. Une boucle crée une répétition<span>
---

- **for n in range (6) :** permet de répéter 6 fois les mêmes instructions
- les instructions qui se répètent sont décalées : **indentée**


```python
terme = 10 # valeur de départ
print(terme) # affiche la valeur de départ

for n in range(6) :  # répétition 6 fois
    terme = terme + 5
    print(terme)
```
