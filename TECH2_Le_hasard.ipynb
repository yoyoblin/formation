>> Simuler une semaine en  ajoutant une boucle

```python
from random import *

for c in range(7) :
  n = int (random()+ 0.7) 
  if n == 1 : print("il pleut") 
  else : print ("il ne pleut pas")
```

>> En ajoutant une fonction :

```python
from random import *

meteo = "" 

def pluie() :
  n = int (random()+ 0.8)
  if n == 1 : return " pluie"
  else : return " soleil"

for c in range (7) :
  meteo = meteo + pluie()

print(meteo)
```

## **2. Exemple avec des listes :** 

#### - Progressivement on apporte de la complexité : Les listes sont des objets "int" , "float" ou "str" placés entre crochets. 

```python
from random import*

univers = ["pile", "face"]
issue = choice(univers)   # tirage au sort dans la liste univers
print (issue)
```

```python
from random import*

univers = ["pile", "face"]

for i in range (10) : # pas besoin de compliquer ici avec la notion d'intervalle
  issue = choice(univers) 
  print (issue)
```

>> Pour obtenir un résultat exploitable il faut créer une liste avec les issues :

```python
from random import*

univers = ["pile", "face"]
tirage = [] # liste vide de résultats

for i in range (10) :
  issue = choice(univers) 
  tirage.append(issue)  # ajout de la face à la liste de tirage

print(tirage)
```

>> Maintenant il s'agit de compter les issues :

```python
from random import*

univers = ["pile", "face"]
tirage = [] # liste vide de résultats

for i in range (10) :
  issue = choice(univers) 
  tirage.append(issue)  

n = tirage.count("face")
f = round(n/10,2)
print ("fréquence de face",f)
```

```python
from random import*

univers = ["pile", "face"]
tirage = [] # liste vide de résultats

t = int(input("nombre de tirages ="))

for i in range (t) :
  issue = choice(univers) 
  tirage.append(issue)  

n = tirage.count("face")
f = round(n/t,2)
print ("fréquence de face",f)
```
