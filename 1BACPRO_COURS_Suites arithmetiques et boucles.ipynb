## <span style="color:orange"> Une boucle `while` répète des instructions "tant que" une condition est vraie. On ne connait pas à l'avance le nombre de répétitions (c'est une boucle non bornée)

### a) Exemple :
> `while n <= 7` : permet de répéter des instructions **tant que** la variable `n` est inférieure ou égale à 7
>- <span style="color:green"> **Modifiez la limite à 10 :**
>- <span style="color:green"> **changez le symbole `<=`en `<`:**

```python
n = 1   # debut de la numerotation

while n <= 7 :  # boucle tant que
    print(n)
    n = n + 1   # numero suivant
```

>- **Remarque 1:** il faut fixer une valeur de départ à la variable `n`
>- <span style="color:red"> **Remarque 2:** il ne faut surtout pas oublier **d'augmenter** la valeur de n à chaque boucle avec `n = n+1` sinon vous créez une boucle infinie...
