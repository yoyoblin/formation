# **PEDA5 Le chiffre de César**

### Le chiffre de César est la méthode de cryptographie la plus ancienne communément admise par l'histoire. Il consiste en une substitution mono-alphabétique : chaque lettre est remplacée par une seule autre, selon un certain décalage dans l'alphabet. César utilisait un décalage de 3 lettres : A devient D, B devient E, C devient F, etc...

> La proposition ci-dessous tient compte des limitations du référentiel de seconde, sans liste et avec des outils simples ...est ce envisageable avec les élèves ? sous forme algorithmique ?

```python
texte = "rfnywjhtwgjfzxzwzsfwgwjujwhmjyjsfnyjsxtsgjhzskwtrfljrfnywjwjsfwiufwqtijzwfqqjhmjqznynsyfujzuwjxhjqfslfljjygtsotzwrtsxnjzwizhtwgjfzvzjatzxjyjxotqnvzjatzxrjxjrgqjegjfzxfsxrjsynwxnatywjwfrfljxjwfuutwyjfatywjuqzrfljatzxjyjxqjumjsncijxmtyjxijhjxgtnxfhjxrtyxqjhtwgjfzsjxjxjsyufxijotnjjyutzwrtsywjwxfgjqqjatncnqtzawjzsqfwljgjhqfnxxjytrgjwxfuwtnjqjwjsfwixjsxfnxnyjyinyrtsgtsrtsxnjzwfuuwjsjevzjytzykqfyyjzwanyfzcijujsxijhjqznvznqjhtzyjhjyyjqjhtsafzygnjszskwtrfljxfsxitzyjqjhtwgjfzmtsyjzcjyhtskzxozwfrfnxzsujzyfwivztssjqduwjsiwfnyuqzx"
message = ""
code = int(input("clé de codage"))

def decod(lettre,code) :
  alphabet = "abcdefghijklmnopqrstuvwxyz"
  rang = alphabet.index (lettre) - code
  return alphabet[rang]

for lettre in texte :
  message = message + decod(lettre,code)

print (message)
```
