### A.1.c. Chronom√©trer le temps d'ex√©cution avec `timeit`

Il est possible de chronom√©trer le temps d'ex√©cution total de `10 000` appels d'une fonction, avec le module `timeit`. Le r√©sultat est donn√© en secondes :

````python
from timeit import timeit

def listeZeros(n):
    lst=[]
    for i in range(n):
        lst.append(0)
    return lst

timeit("listeZeros(500)", globals=globals(), number= 10_000)
````

### A.1.d. Visualiser un graphique avec `matplotlib.pyplot`

Il est possible de visualiser un graphique avec `matplotlib.pyplot`.

Par exemple, apr√®s avoir ex√©cut√© la cellule pr√©c√©dente, on peut comprendre ci-dessous la complexit√© en O(n) de l'algorithme :

````python
import matplotlib.pyplot as plt

x = list(range(0,50,1))

y = []
for i in x:
    t = timeit("listeZeros("+str(i)+")", globals=globals(), number= 10_000)
    y.append(t)

plt.clf()
plt.title("Temps d'ex√©cution en fonction du param√®tre")
plt.plot(x, y)
plt.xlabel('taille de la liste')
plt.ylabel("temps d'ex√©cution (en s)")
plt.show()
````

### A.1.f. Modifier une image avec `PIL`

Il est possible de modifier une image avec `PIL` : 

````python
from PIL import Image
monImage1=Image.open("buse.jpg")#Cette image a une dimension de 400√ó200 pixels
monImage2=Image.new('RGB',(400,200))#Cr√©e une seconde a une dimension de 400√ó200 pixels

for x in range(400):
    for y in range(200):
        r,v,b = monImage1.getpixel((x, y)) #r,v,b prennent les valeurs des composante RVB du pixel (x,y) de Image1
        g=(r+v+b)//3                 #calculer ici la valeur de g                  
        monImage2.putpixel( (x, y),(g,g,g))#On fixe la couleur du pxel (x,y) de Image2

monImage4=Image.new("RGB",(806,204))
monImage4.paste(monImage1, (2, 2)  )
monImage4.paste(monImage2, (404, 2))
monImage4                #On affiche une image compos√©e de Image1 et Image2
````
