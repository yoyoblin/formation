## **1. Découverte des fonctions python**


- **Une fonction** en langage python est **une partie d’un programme que l’on « appelle »** afin d’automatiser une série d'instructions et qui comporte un plusieurs paramètres.
- La fonction **renvoie (return)** des valeurs. (Une fonction qui ne renvoie rien est une procédure, un sous programme).
- Une fonction python peut être utile par exemple dans une boucle ou quand on doit **souvent utiliser les mêmes instructions.** Celles ci sont identées.
- Une fonction doit être créée **avant** la partie du programme qui l'appelle

> Exemple avec un seul paramètre : il est plus simple que le paramètre soit nommé de manière identique, mais ça marche aussi si on le change.

```python
def fonction(n) : # la fonction
  return n*"bonjour "

print (fonction(3)) # la partie principale du programme appelle la fonction
```

```python
def fonction(i) : # Le paramètre avec une variable de la fonction qui peut être différente
  return i*"bonjour "

n = 3
print(fonction(n))
```

> Exemple avec plusieurs paramètres :

```python
def fonction(a,b,c) :
  return (a + b)*c

print (fonction(2,3,4))
```

> **Attention !** les variables utilisées dans les fonctions ne sont pas utilisables en dehors de celles ci : il y a une zone mémoire séparée, même si les varables ont le même nom comme ci-dessous.

```python
y = 0 

def f(x) :
  y = x**2
  return x**2

print (f(2)) # affiche le retour de la fonction
print (y)    # ne tient aucun compte de la valeur affectée dans la fonction
```

> Exemple avec des chaînes de caractères :


```python
def pourcent(lettre,mot) : # fonction pourcentage de lettre dans un mot
  n = mot.count(lettre)
  l = len(mot)
  p = round(n/l*100)
  return "dans "+mot+" la fréquence de "+str(lettre)+" est "+str(p)+"%"

print(pourcent("a","ananas"))
print(pourcent("b","baobab"))
```

```python
def pourcent(lettre,mot) : # fonction pourcentage de lettre dans un mot
  n = mot.count(lettre)
  l = len(mot)
  p = round(n/l*100)
  return "dans "+mot+" la fréquence de "+str(lettre)+" est "+str(p)+"%"

mot = input("entrez un mot ")
for c in mot :
  print(pourcent(c,mot))
```

> Fonction affine par morceau

```python
def f(x):   
  if x <= 4 :
    return 2*x
    
  else :
    return x+4

for x in range (0,8) :
	print("f(",x,")=",f(x))
```

> Comparaison de deux fonctions :

```python
def f(x) :
  return 2*x

def g(x) :
  return x+4

x =0
while f(x) <= g(x) :
  print(x," ",f(x)," ",g(x))
  x = x + 0.5
```
