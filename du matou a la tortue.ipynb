### **2. Turtle : les boucles**



- **La boucle for** avec une variable de boucle. 
C'est une boucle bornée : Elle répète les instructions un nombre prédéfini de fois.

- **La boucle while** avec une variable incrémentée. 
C'est une boucle non bornée : Elle répète les instructions tant que la condition est vraie.



### - boucle "for"
**Dans les exemples ci-dessous la boucle est bornée :** la variable de boucle prend toutes les valeurs (int ou str) de d'un intervalle ou de la liste. 

On l'utilise ou pas dans les instructions de la boucle, car ce n'est pas une obligation. De plus et elle n'a pas besoin d'être incrémentée.


```python
from turtle import *
speed(6) 

for i in range(6) :
  forward(100)
  left(60)

done()
```

```python
from turtle import *
speed(6) 

for i in ["red","green","blue","orange","yellow","grey"] :
  color(i)
  forward(100)
  left(60)

done()
```

### - boucle "while"
**Dans les exemples ci-dessous la boucle est non bornée :** elle se répète "tant que la condition est vraie", de plus elle a besoin d'être incrémentée 

```python
from turtle import *
speed(6) 

i = 0 # initialisation obligatoire de la variable compteur
while i < 6 :
  forward(100)
  left(60)
  i = i + 1

done()
```

### **3. Turtle : les fonctions**


- **Une fonction** renvoie une variable par l'intermédiaire de **return**
- **Une fonction sans renvoi** est une procédure, un sous programme.



```python
from turtle import *
from random import *
speed(4)

def d(i) :  # cette fonction a un parametre et renvoie une valeur
  return i*10

def a() :  # cette fonction n'a pas de parametre et renvoie une valeur
  return randint(50,100)

for i in range (0,20) :
  forward(d(i))
  left(a())

done()
```

```python
from turtle import *
from random import *
speed(8)

def move1(n) : # cette fonction ne renvoie rien
  forward(n) 
  left(60)

def move2(n) : # cette fonction ne renvoie rien non plus
  forward(n) 
  right(60)

for i in range (0,50) :
  n = randint(0,50)
  if n < 25 :
    move1(n)
  else :
    move2(n)

done()
```
