### **1. Utilisation des boucles en algorithmique**


---





Les structures itératives permettent d’exécuter plusieurs fois un bloc d’opérations.

- **Boucles bornées (répéter n fois) :** La boucle « FOR » permet de répéter un bloc d’opérations un nombre de fois défini en faisant varier automatiquement la valeur d’une variable (compteur ou variable de boucle).



*   La variable **"compteur"** est créée automatiquement, elle n'a pas besoin d'être initialisée et incrémentée. Le pas est 1 par défaut dans python
*   **Il n'y a pas dans le langage python de boucle "répéter n fois"** contrairement à scratch, nous pouvons utiliser une boucle de type "for" qui opère entre deux valeurs et c'est l'occasion de rappeler la notion d'intervalle en math.

- **Boucles non bornée :** « WHILE » permet de répéter un bloc d’opérations **tant qu’une condition** est satisfaite. Il faut être attentif au problème de la terminaison et ne pas créer de boucle infinie.
- **La boucle "while"** est aussi moins facile d'usage car il faut en amont, créer les variables qui sont utilisées dans la boucle.

### **2. Introduction progressive des boucles : la boucle FOR**


---




- **la variable compteur** prend toutes les valeurs d'une liste ou d'une autre variable. 


*   La boucle commence par **FOR** et se termine par **deux points :**
*   Toutes les instructions **indentées (décalées)** font partie de la boucle.
*   La variable **"compteur" est créée automatiquement** et d'ailleurs n'est pas forcément utilisée dans la boucle.








> Ci-dessous on affiche la variable compteur qui "balaye" toutes les lettres d'un mot et on affiche ce mot. Mais vous pouvez ne pas tenir compte de  la variable (exemple 2)

```python
# exemple 1
for i in "crocodile" :
  print(i)
```

```python
# exemple 2
for l in "puce" :
  print("hop")
```

> Dans l'exemple ci-dessous la variable prend toutes les valeurs de la liste **range(4)** qui donne les entiers 0, 1, 2, 3

```python
for x in range(4) :  # l'instruction range(n) crée une liste de valeurs de 0 à n-1
  print (x)
  # Modifiez le programme pour aller jusqu'à 10
```

```python
for x in range (0,4) : # c'est la même chose mais c'est plus explicite cela correspond à l'intervalle [0;4[
  print (x)
  # Modifiez le programme pour afficher les entiers de 2 à 6
```

```python
for x in range (2,12,2) :  # on peut avoir un pas entier, mais non décimal
  print(x)
```

> Pour résumer :
>>-  for c in .....crée **une variable compteur**
>>-   L'instruction **range(a,b)** crée une liste de valeurs entières entre a et b-1
>>-  L'instruction **range(a)** crée une liste de valeurs entières entre 0 et a-1
>>-   On peut tout de même avoir un pas différent de 1, mais **entier**.



> **Il est possible d'utiliser un pas décimal**, mais il faut utiliser un module mathématique comme numpy, ou passer aux boucles while

```python
from numpy import *
for n in arange(0,1,0.2) :
  print (n)
```

### **3. La boucle WHILE**


---






*   **A la différence de la boucle "for"** elle n'a pas de borne et peut être exécutée **indéfiniment**
*   **Attention !!!** si vous executez le code ci-dessous il faudra mettre en pause très rapidement !!!! aussi **je n'ai pas écrit les deux points :**faites le test dans votre interprèteur python....



```python
while True # erreur volontaire : cette boucle est infinie
  print("La vérité est toujours vraie")
```

**Quelques indications :**

>- Vous devez **créer et initialiser** la variable compteur avant la boucle
>- Vous devez penser à **incrémenter cette variable** à la fin des instructions de la boucle, ou au début... au choix

>- **Les opérateurs de comparaison** sont : ==; !=, <= etc
>- **La condition peut être une iniégalité ou même un encadrement** a < x < b. C'est l'occasion de rappeler les inégalités, les encadrements...


```python
x = 0  # on crée et initialise la variable de la boucle

while x < 5:
  print (x,"-->",x**2)


# ça ne marche pas ? devinez pourquoi...
# !!!! mais surtout stoppez l'exécution de la cellule avant !!!
```

> Il peut être interessant de fixer un objectif à une suite numérique : "à partir de quelle rang n, le terme Un dépasse t'il la valeur 90 ?"


```python
u = 3 # premier terme de la suite
r = 5.5 # raison
n = 1 # indice de la suite

while u < 90 : 
  print("U",n,"=",u)
  u = u + r  
  n = n + 1

# toutes les valeurs affichées c'est envahissant... modifiez le programme pour ne plus afficher que la dernière valeur
```

> Résolution par balayage d'une inéquation de type : $$f(x)<a$$



```python
x = 0
y = 0 # il faut la créer même si la valeur initiale ne sert à rien

while y <= 5 :
  y = x**2
  print (x,"-->",y) 
  x = x + 0.5  # incrément
  # Modifiez l'incrément pour affiner la résolution
  # remarque : on dépasse l'objectif... pourquoi ?
```

> Deux possibilités pour ne pas dépasser l'objectif : 


>>1.   Jouer avec **l'ordre des intructions** (travail nécessaire sur l'algorithmique en amont).
>>2.   Utiliser l'instruction **Break**





```python
x = 0
y = 2 # erreur volontaire

while y <= 5 :
  print (x,"-->",y) 
  x = x + 0.5
  y = x**2
# La valeur au départ est correcte si on l'écrit correctement, mais on triche
```

```python
x = 0
y = 0 

while y <= 5 :
  y = x**2
  if y > 5 : break # on peut aussi résoudre en partie la problèmatique précédente avec l'instruction break
  print (x,"-->",y) 
  x = x + 0.5
```
