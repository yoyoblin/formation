## <font color="red" > 3. Chaînes de caractères : des chiffres et des lettres </font> 

- Les lettres sont appelées **"chaînes de caractères"**,en anglais **string** et en raccourci **str**.
- On définit une chaîne de caractères avec des doubles `"chaîne"`ou des simples `'chaîne'`

```python
mot1 = "armoire"
mot2 = 'gruyère'
print (mot1,"à",mot2)
```


- **On définit une chaîne de caractères avec des doubles `"chaîne"`ou des simples `'chaîne'`**
- On peut faire des **opérations** avec des chaînes de caractères : on appelle ça des **concaténations**.

```python
# Pour concaténer des chaines on les ajoute avec +
a = "coul"
b = "euvre"
c = a + b
print (c)
c = c + " " +"verte" # ajout d'un espace vide
print (c)
```

```python
a = "HA!"
b = "HO!"
lol = 5*(2*a + b)  # La concaténation peut utiliser les opérateurs +, * et () pour créer des combinaisons
print(lol)
```

- **Compter le nombre de caractères** dans une chaîne avec `len( )`
- **Compter le nombre d'occurences** d'une lettre dans un mot avec `.count( )`

```python
alphabet ="abcdefghijklmnopqrstuvwxyz"
n = len(alphabet) # longueur de la chaîne de caractères
print(n)
```

```python
alphabet ="abcde fghijklmnopqrstuvwxyz" # un espace est considéré comme un caractère
n = len(alphabet) 
print(n) 
```

```python
mot ="anticonstitutionnel"
n = mot.count("n") # compter le nombre de lettres n
print(n)
```

##  <font color="red" > 5. Les listes </font> 

- **Les listes** permettent de stocker plusieurs résultats dans un ordre donné entre crochets [ ]
- On peut mélanger plusieurs types dans une liste : des nombres, des chaînes...
- On peut **compter** le nombre d'éléments dans une liste avec len( ).
- On peut **chercher** un élément par son index (numéro) avec liste [ ]
- **Important :** Les index commencent à 0
- **On ajoute un élément** dans une liste avec liste.append()

> ###  - <font color="green" > Le plus utile :

```python
# Créer des listes et les additionner
liste1 =[1,"boa",33,"a"]
liste2 =[2,"python",42]
print(liste1 + liste2)
```

```python
# Concaténer des listes
liste1 =[1,"boa",33,"a"]
liste2 =[2,"python",42]
print(liste1 + 2*liste2)
```

```python
# Pour compter le nombre d'éléments dans une liste
liste =[1,"boa",33,"a"]
print(len(liste))
```

```python
# Pour compter le nombre d'occurences d'un élément dans une liste
liste =[1,7,2,3,4,4,6,5,6,6,7,1,8,7]
n = liste.count(7)
print(n)
```

```python
# l'élément d'index 2 est ici le troisième
liste =[1,"boa",33,"a"]
print(liste[2])
```

```python
# Pour ajouter le nombre 42 à la liste
liste =[1,"boa",33,"a"]
liste.append(42)
print(liste)
```

```python
# Pour additionner tous les éléments dans une liste
liste =[1,7,2,3,4,4,6,5,6,6,7,1,8,7]
s = sum(liste)
print(s)
```

```python
# Pour trouver l'index d'un élément dans une liste (le premier s'il y en a plusieurs)
liste =[1,9,3,5,4,4,5,6,7,6,8,10,11]
n = liste.index(6)
print(n)
```

```python
# Pour ordonner une liste (ordre croissant) sans modifier la liste d'origine.
liste1 =[1,9,3,5,4,4,5,6,7,6,8,10,11]
liste2 = sorted(liste1) # attention ce n'est pas une variable !!!
print(liste1)
print(liste2)
```

```python
# Pour ordonner une liste (ordre croissant) en modifiant la liste d'origine
liste =[1,9,3,5,4,4,5,6,7,6,8,10,11]
liste.sort() # attention ce n'est pas une variable !!!
print(liste)
```

```python
# Pour ordonner une liste (ordre décroissant) en modifiant la liste d'origine
liste =[1,9,3,5,4,4,5,6,7,6,8,10,11]
liste.reverse() # attention ce n'est pas une variable !!!
print(liste)
```

```python
# Créer une liste d'entiers impairs
liste =list(range(1,100,2))
print(liste)
```

```python
# Créer une liste de nombres décimaux
import numpy
liste =list(numpy.arange(1,10,0.2))
print(liste)
```

> ###  - <font color="green" > Listes par compréhension : permet d'inclure des boucles et des conditions dans une liste

```python
# générer une liste d'entiers de 1 à ...
liste =[n for n in range(1,20) ]
print(liste)
```

```python
# Jouer aux dés...
from random import*
alea = [randint(0,6) for n in range(20)] 
print(alea)
```

```python
# Générer des nombres sous conditions
impairs =[n for n in range(1,50) if n%2==1 ]
print(impairs)
```

```python
# tableau de valeurs pour une fonction
valeurs_f =[x**2 for x in range(1,100) if 30 <x<40 or x%2 == 1]
print(valeurs_f)
```

```python
# tableau de valeurs pour une fonction
valeurs_f =[25- x**2 if 25- x**2 >=0 else 'négatif' for x in range(1,10)]
print(valeurs_f)
```
