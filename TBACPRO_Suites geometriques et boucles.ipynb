## 1. Rappels sur les boucles `for` et `while`

### <span style="color:orange"> Une boucle `for` crée un nombre défini de répétitions (c'est une boucle bornée)

> `for n in rang(1,10)` : permet de répéter **9 fois** les instructions


```python
u = 15 
for n in range(1,10) : # crée 9 répétitions
    print(u)
    u = u*2 
```

### <span style="color:orange"> Une boucle `while` répète des instructions "tant que" une condition est vraie. On ne connait pas à l'avance le nombre de répétitions (c'est une boucle non bornée)

> `while u <= 100` : permet de répéter des instructions **tant que** la variable `u` est inférieure ou égale à 100

```python
u = 15 
while u <= 100 :  # boucle tant que
    print(u)
    u = u*2 
```
